### Track Name

- Below we have found the 10 songs with the highest popularity level. But the two codes gave us different results.
- The reason for this is that we only used the song as a basis in the second code.
- In the first code, we took this song and the artist who performed it as a basis.
- In other words, more than one artist may have performed a song and this may have affected its popularity.
- This may cause us to see different popularity values when there are versions of songs with the same name by different artists.

```python
track_popularity_top10 = df.groupby(['track_name', 'track_artist'])['track_popularity'].mean().sort_values(ascending=False)
print("Top 10 track with the highest average Track popularity : ")
track_popularity_top10.head(10)
```

```python
track_popularity_last10 = df.groupby(['track_name', 'track_artist'])['track_popularity'].mean().sort_values(ascending=True)
print("Last 10 track with the lowest average Track popularity : ")

# Choose at least 10 tracks with non-zero scores
non_zero_popularity_last10 = track_popularity_last10[track_popularity_last10 > 0].head(10)
non_zero_popularity_last10.head(10)
```

#### Trevor Daniel's popularity changes over the years

```python
# Convert 'track_album_release_date' column to datetime format, fill in incorrect or missing values with 'NaT' (Not a Time)
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce')

# Remove rows with NaN (Not a Number) or 'NaT' values
df = df.dropna(subset=['track_album_release_date'])

# Group Queen's popularity values by year
Queen_popularity_by_year = df[df['track_artist'] == 'Queen'].groupby(df['track_album_release_date'].dt.year)['track_popularity'].mean()
Queen_popularity_by_year
```

```python
# Group Queen's popularity values by year
Queen_popularity_by_year = df[df['track_artist'] == 'Queen'].groupby(df['track_album_release_date'].dt.year)['track_popularity'].mean()


plt.figure(figsize=(10, 6))
Queen_popularity_by_year.plot(marker='o')
plt.title('Queen\'s Popularity - Change Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(True)
plt.show()
```

### Playlist Genre

```python
print(f"We have {len(df['playlist_genre'].unique())} different type of playlist. Such as : ")
print(df['playlist_genre'].unique())
```

```python
playlist_popularity = df.groupby('playlist_genre')['track_popularity'].mean().sort_values(ascending=False)
print("The list of playlist genre with the highest average track popularity : ")
playlist_popularity.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=playlist_popularity.index, y=playlist_popularity.values, hue=playlist_popularity.index, palette='viridis')
plt.title('Average Popularity Values of Playlist Genres')
plt.xlabel('Playlist Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Playlist Genre', bbox_to_anchor=(1, 1))
plt.show()
```

#### Distribution of Track Popularity Across Playlist Genres

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['playlist_genre'], y=df['track_popularity'])
plt.title('Playlist Genre vs. Track Popularity')
plt.xlabel('Playlist Genre')
plt.ylabel('Track Popularity')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

#### Relationship Between Album Release Date and Track Popularity (Pop Remix Playlist)

```python
# Select only data for playlist "Pop Remix"
pop_remix_df = df[df['playlist_name'] == 'Pop Remix']

# Draw the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(pop_remix_df['track_album_release_date'], pop_remix_df['track_popularity'], alpha=0.5)
plt.title('Album Release Date vs. Track Popularity (Pop Remix Playlist)')
plt.xlabel('Album Release Date')
plt.ylabel('Track Popularity')
plt.show()
```

### Relationship Between Popularity and Features

- Columns to Use: `track_popularity`, `danceability`, `energy`, `valence`

```python
correlation_matrix = df[['track_popularity', 'danceability', 'energy', 'valence']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Relationship Between Popularity and Features')
plt.show()
```

### The corelation for all the numerique columns 

- Loudness and energy have the highest correlation. However, the correlation between acousticness and energy is the lowest.

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Create the correlation matrix
correlation_matrix = numeric_columns.corr()

# Draw the heat map
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Between All Numerical Columns')
plt.show()
```

### Time Analysis

- Columns to use: `track_popularity`, `track_album_release_date`, `tempo`, `energy`
- Description: Examine changes in popularity by album release date, identifying trends in music characteristics such as tempo and energy over time.

```python
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce')
```

```python
# Remove rows with NaN (Not a Number) or 'NaT' values
df = df.dropna(subset=['track_album_release_date'])
```

```python
# Get the years in the 'track_album_release_date' column as unique and sort them in descending order
sorted_unique_years = sorted(df['track_album_release_date'].dt.year.unique(), reverse=True)

print("Unique Years:")
print(sorted_unique_years)
```

```python
# Plot the relationship between album release date and track popularity
plt.figure(figsize=(12, 6))
plt.scatter(df['track_album_release_date'], df['track_popularity'], alpha=0.5)
plt.title('Album Release Date vs. Track Popularity')
plt.xlabel('Album Release Date')
plt.ylabel('Track Popularity')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['track_album_release_date'], df['tempo'], label='Tempo', alpha=0.7)
# plt.scatter(df['track_album_release_date'], df['energy'], label='Energy', alpha=0.7)
plt.title('Tempo Change Over Time')
plt.xlabel('Album Release Date')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
# plt.scatter(df['track_album_release_date'], df['tempo'], label='Tempo', alpha=0.7)
plt.scatter(df['track_album_release_date'], df['energy'], label='Energy', alpha=0.7)
plt.title('Energy Change Over Time')
plt.xlabel('Album Release Date')
plt.ylabel('Values')
plt.legend()
plt.show()
```
